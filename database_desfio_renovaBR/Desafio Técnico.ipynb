Tratamento de dados - perfil_eleitorado_2020.csv

```python
import pandas as pd

# Carregar o arquivo CSV usando tabulação como separador
file_path = 'D:\\gusta\\database_desfio_renovaBR\\perfil_eleitorado_2020.csv'
df_perfil = pd.read_csv(file_path, sep=';', encoding='ISO-8859-1')

# Selecionar apenas as colunas relevantes para a análise
colunas_relevantes = ["ANO_ELEICAO", "SG_UF", "CD_MUNICIPIO", "NM_MUNICIPIO", "NR_ZONA",
                "DS_GENERO", "DS_ESTADO_CIVIL", "DS_FAIXA_ETARIA", "DS_GRAU_ESCOLARIDADE",
                "QT_ELEITORES_PERFIL", "QT_ELEITORES_BIOMETRIA"]
df_perfil = df_perfil[colunas_relevantes]


```

```python
# Remover "#" do início e do final de "#NULO#"
colunas_a_verificar = ['NM_PARTIDO', 'DS_AGREGADAS', 'SG_PARTIDO']
for coluna in colunas_a_verificar:
    if coluna in df_turno.columns:
        df_turno[coluna] = df_turno[coluna].str.replace('^#|#$', '', regex=True)
```
